```python
import numpy as np
import matplotlib.pyplot as plt
height=np.random.normal(140,20,1000)
plt.figure(dpi=200)
plt.title("Height Distribution")
plt.xlabel("Height(cm)")
plt.ylabel("Frequency")
plt.hist(height,bins=30,ec="gold")
plt.show()
```

**Difference between interpreter and compiler:**

      A Compiler takes a program as a whole and then generate its intermediate machine codes wheraas an Interpreter takes
      single lines of a code, it never generate intermediate machine codes.
      Compiler display all the errors after compiling the whole program together at the same time but Interpreter displays
      the errors of every single line one by one.
